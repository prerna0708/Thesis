```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
import os
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
directory = os.getcwd()
path = directory + '/Data/'
```

## Reading files

```python
Centrality=pd.read_csv(path + 'df_final_graphoptions.csv')
Centrality_wo_options=pd.read_csv(path + 'df_final_graph_wo_options.csv')
```

## Calculating Call Put ratio

```python
Centrality['Degree_reverse_wt_put'] = Centrality['Degree_reverse_wt_put'].replace([0],1)
Centrality['Degree_reverse_wt_call'] = Centrality['Degree_reverse_wt_call'].replace([0],1)
```

```python
Centrality["call_put_ratio"]=(Centrality['Degree_reverse_wt_call'])/(Centrality['Degree_reverse_wt_put'])
Centrality.replace([np.inf, -np.inf], 0, inplace=True)
```

## Sorting based on Degree_natural_wt

```python
Centrality=Centrality.sort_values(by=['Degree_natural_wt'], ascending=False).reset_index(drop=True)
df=Centrality.iloc[0:15]
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10),  sharey=True)

sns.barplot(ax=axes[0,0], x="Degree_natural_wt", y="name", data=df,
                 color="salmon", saturation=.5)
axes[0,0].set_title('Degree_natural_wt')

sns.barplot(ax=axes[0,1], x="Degree_reverse_wt", y="name", data=df,
                 color="salmon", saturation=.5)
axes[0,1].set_title('Degree_reverse_wt')

sns.barplot(ax=axes[0,2], x="Page_Rank_wt_natural", y="name", data=df,
                 color="salmon", saturation=.5)
axes[0,2].set_title('Page_Rank_wt_natural')

sns.barplot(ax=axes[1,0], x="Closeness_natural", y="name", data=df,
                 color="salmon", saturation=.5)
axes[1,0].set_title('Closeness_natural')

sns.barplot(ax=axes[1,1], x="Betweeness", y="name", data=df,
                 color="salmon", saturation=.5)
axes[1,1].set_title('Betweeness')

sns.barplot(ax=axes[1,1], x="Betweeness", y="name", data=df,
                 color="salmon", saturation=.5)
axes[1,1].set_title('Betweeness')

sns.barplot(ax=axes[1,2], x="call_put_ratio", y="name", data=df,
                 color="salmon", saturation=.5)
axes[1,2].set_title('Highest Degree_natural_wt')
```
