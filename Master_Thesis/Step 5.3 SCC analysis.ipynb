```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
import os
```

```python
directory = os.getcwd()
path = directory + '/Data/'
```

## Reading files

```python
Centrality=pd.read_csv(path + 'df_final_graph_wo_options.csv')
scc=pd.read_csv(path + 'scc.csv')
```

## Centrality of scc nodes

```python
scc_centrality=Centrality[Centrality['cik'].isin(scc['id'])]
```

## Finding nodes with high pagerank within SCC

```python
Centrality=scc_centrality.sort_values(by=['Page_Rank_wt_natural'], ascending=False).reset_index(drop=True)
Centrality['id_od']=Centrality["Degree_reverse_wt"]-Centrality["Degree_natural_wt"]
df=Centrality.iloc[0:19]
fig, axes = plt.subplots(2, 2, figsize=(20, 10),  sharey=True)

sns.barplot(ax=axes[0,0], x="id_od", y="name", data=df,
                 color="salmon", saturation=.5)
axes[0,0].set_title('indegree_outdegree')

sns.barplot(ax=axes[0,1], x="Page_Rank_wt_natural", y="name", data=df,
                 color="salmon", saturation=.5)
axes[0,1].set_title('Page_Rank_wt_natural')

sns.barplot(ax=axes[1,0], x="Closeness_natural", y="name", data=df,
                 color="salmon", saturation=.5)
axes[1,0].set_title('Closeness_natural')

sns.barplot(ax=axes[1,1], x="Betweeness", y="name", data=df,
                 color="salmon", saturation=.5)
axes[1,1].set_title('Betweeness')
```
